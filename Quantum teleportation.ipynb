```python
from qiskit import *
```

```python
qc = QuantumCircuit(3,3)
```

```python
%matplotlib inline
qc.draw(output = 'mpl')
```

```python
qc.z(0)
qc.barrier()
qc.draw('mpl')
```

```python
qc.h(1)
qc.cx(1,2)
```

```python
qc.draw('mpl')
```

```python
qc.cx(0,1)
qc.h(0)
```

```python
qc.draw('mpl')
```

```python
qc.barrier()
qc.measure([0,1],[0,1])
qc.draw('mpl')
```

```python
qc.barrier()
qc.cx(1,2)
qc.cz(0,2)
qc.draw('mpl')
```

```python
qc.measure(2,2)
qc.draw('mpl')
simulator = Aer.get_backend('qasm_simulator')
result = execute(qc, backend = simulator,shots = 2048).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
